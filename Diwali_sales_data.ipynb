```python
df = pd.read_csv('/content/Diwali Sales Data.csv' , encoding = 'unicode_escape')
df.head()
```

```python
df.info()
```

```python
df.drop(['Status' , 'unnamed1' ], inplace = True , axis = 1)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
#Missing data in terms of percentage
df.isnull().sum() * 100 / len(df)
```

```python
# dropna to drop the missing values
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
df['Amount'] = df['Amount'].astype('int')  # Updating the Existing Column
```

```python
df
```

```python
# Adding a column with respect to dollars
df['USD'] = df['Amount'] / 84.11
```

```python
# Renaming the column Marital_Status >> Shadi
df.rename(columns = {'Marital_Status' : 'Shadi'})
```

```python
# Plot a bar graph over gender
sns.countplot(data = df , x = 'Gender' , palette = 'viridis')
plt.show()
```

```python
df['Gender'].value_counts().plot(kind = 'bar')
```

# Documentation

From the above graph we can see that females are purchasing more compare to male

```python
# I want to see who is spending more amount female or male
sales_gen = df.groupby('Gender')['Amount'].sum().reset_index()
sales_gen
```

```python
sns.barplot(data = sales_gen , x = 'Gender' , y = 'Amount')
plt.show()
```

```python
# state
df['State'].value_counts().head(10)
```

```python
df.sample()
```

```python
# Orders placed w.r.t. the state
sales_state_ord = df.groupby('State')['Orders'].sum().sort_values(ascending = False).reset_index()
```

```python
sns.barplot(data = sales_state_ord , x = 'State' , y = 'Orders' , palette = 'viridis')
plt.xticks(rotation = 90)
plt.show()
```
